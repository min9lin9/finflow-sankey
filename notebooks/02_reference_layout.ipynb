# FinFlow Sankey: Reference Layout

The reference layout shows detailed account nodes. Use it when you have individual revenue/expense accounts.

```
import polars as pl
from finflow_sankey import FinancialSankey
```

## 1. Prepare detailed data

```
df = pl.DataFrame({
    "account": ["Product Sales", "Service Sales", "Cost of Sales", "Payroll", "Rent", "Transport", "Net Income"],
    "value": [156.0, 73.0, -163.0, -18.0, -9.0, -5.0, 34.0],
    "period": ["FY2025"] * 7,
    "currency": ["KRW"] * 7,
    "statement": ["income_statement"] * 7,
    "section": ["revenue", "revenue", "cost_of_revenue", "operating_expenses", "operating_expenses", "operating_expenses", "profit"],
})
df
```

## 2. Render reference layout

```
fig = (
    FinancialSankey.income_statement(df, layout="reference")
    .validate()
    .render(title="Detailed Income Statement (Reference Layout)")
)
fig.show()
```

## 3. Use a dark theme

```
fig_dark = (
    FinancialSankey.income_statement(df, layout="reference")
    .validate()
    .render(title="Detailed Income Statement (Dark)", theme="dark")
)
fig_dark.show()
```
